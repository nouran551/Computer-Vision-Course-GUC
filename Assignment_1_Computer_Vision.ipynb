```python
def displayImages(imgs):
  r = len(imgs) //2
  if len(imgs) % 2 != 0:
    r+=1
  plt.subplot(nrows=r,ncols=2,figsize=(20,20))
  for i in range (0, len(imgs)):
    plt.subplot(r,2,i+1)
    plt.imshow(imgs[i],cmap = 'gray')
```

```python
image1=cv2.imread('/content/Input1.png',cv2.IMREAD_COLOR)
image2=cv2.imread('/content/Input2.png',cv2.IMREAD_COLOR)

def retrieveImage(image1, image2):
  s = {0.1, 0.5, 0.01, 0.02}
  for i in s :
    bfr_s =cv2.subtract(image1, image2)
    aftr_s = bfr_s * i
    test = cv2.imwrite('image_test_fun2_'+str(i) +'.png' ,aftr_s)
```

```python
retrieveImage(image1, image2)
```
