```python
def displayImages(imgs):
  r = len(imgs) //2
  if len(imgs) % 2 != 0:
    r+=1
  plt.subplot(nrows=r,ncols=2,figsize=(20,20))
  for i in range (0, len(imgs)):
    plt.subplot(r,2,i+1)
    plt.imshow(imgs[i],cmap = 'gray')
```

```python
def scaling42(image1):
  down_width = 128
  down_height = 128
  down_points = (down_width, down_height)
  resized_down = cv2.resize(image1, down_points, interpolation= cv2.INTER_LINEAR)
  test = cv2.imwrite('image_test_fun3.png' ,resized_down)
```
